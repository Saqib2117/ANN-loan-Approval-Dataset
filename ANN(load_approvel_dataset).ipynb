### Data Preprocessing

```python
df.drop('loan_id', axis=1, inplace=True)
```

```python
df.education = df.education.str.strip()
```

```python
df.self_employed = df.self_employed.str.strip()
```

```python
df.loan_status = df.loan_status.str.strip()
```

```python
df.education = df.education.apply(lambda x: 1 if x == 'Graduate' else 0)
df.self_employed = df.self_employed.apply(lambda x: 1 if x == 'Yes' else 0)
df.loan_status = df.loan_status.apply(lambda x: 1 if x == 'Approved' else 0)
```

### Features Selection from the data

```python
df.corr()
```

### Spliting Features and Target

```python
X = df.drop('loan_status', axis=1)
y = df.loan_status
```

### Applying Standard Scalar on Features

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Preparing Training and Testing Data

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.20, random_state=42)
```

### Model Arhitecture

```python
model = Sequential([
    layers.Dense(64, activation='relu'),
    layers.Dense(16, activation='relu'),
    layers.Dense(8, activation='relu'),
    layers.Dense(2, activation='softmax')
])
```

### Compiling the Model

```python
model.compile(
    optimizer='rmsprop',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

### Training the Model on training data

```python
model.fit(X_train, y_train, epochs=50, batch_size=32)
```

### Now testing the Model and Give it Testing Data

```python
model.evaluate(X_test, y_test)
```

```python
model.summary()
```

### Making Function to predict the loan approval based on user input

```python
def predict_loan_status(no_of_dependents, education, self_employed, income_annum,
                        loan_amount, loan_term, cibil_score, residential_assets_value,
                        commercial_assets_value, luxury_assets_value, bank_asset_value):
    
    education = 1 if education.strip().capitalize() == 'Graduate' else 0
    self_employed = 1 if self_employed.strip().capitalize() == 'Yes' else 0
    
    features = np.array([[no_of_dependents, education, self_employed, income_annum,
                          loan_amount, loan_term, cibil_score, residential_assets_value,
                          commercial_assets_value, luxury_assets_value, bank_asset_value]])
    
    features_scaled = scaler.transform(features)
    
    prediction = model.predict(features_scaled)
    predicted_class = prediction.argmax(axis=1)[0]
    confidence = prediction[0][predicted_class]
    
    if predicted_class == 1:
        print(f'Loan Approved — Confidence: {confidence*100:.2f}%')
    else:
        print(f'Loan Rejected — Confidence: {confidence*100:.2f}%')
```

```python
no_of_dependents          = int(input('Enter number of dependents: '))
education                 = input('Education (Graduate/Not Graduate): ')
self_employed             = input('Self Employed (Yes/No): ')
income_annum              = int(input('Enter annual income: '))
loan_amount               = int(input('Enter loan amount: '))
loan_term                 = int(input('Enter loan term (years): '))
cibil_score               = int(input('Enter CIBIL score (300-900): '))
residential_assets_value  = int(input('Enter residential assets value: '))
commercial_assets_value   = int(input('Enter commercial assets value: '))
luxury_assets_value       = int(input('Enter luxury assets value: '))
bank_asset_value          = int(input('Enter bank asset value: '))

predict_loan_status(no_of_dependents, education, self_employed, income_annum,
                    loan_amount, loan_term, cibil_score, residential_assets_value,
                    commercial_assets_value, luxury_assets_value, bank_asset_value)
```
